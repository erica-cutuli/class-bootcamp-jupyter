```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("soggetti-guariti.csv")
df
```

```python
plt.bar(df["eta"], df["guariti_senza_somm"])
plt.show()
```

```python
#df = df.loc[:, ["Date", "A"]]
#df = df.sort_values(by="guariti_senza_somm")
plt.plot(df["eta"], df["guariti_senza_somm"], marker="o", linestyle="-", color="b", label="Value")
plt.show()
```

```python
import numpy as np

# Larghezza delle barre
width = 0.4  

# Generiamo gli indici per l'asse X basati sulle stringhe di etÃ 
x = np.arange(len(df["eta"]))

# Creazione del grafico con barre affiancate
plt.bar(x - width/2, df["guariti_senza_somm"], width=width, label="Senza Somm.")
plt.bar(x + width/2, df["guariti_post_somm"], width=width, label="Post Somm.")

# Etichette testuali sull'asse X
plt.xticks(x, df["eta"], rotation=45)  # Ruotiamo se serve

# Aggiungiamo legenda e titoli
plt.legend()
plt.xlabel("Fascia di EtÃ ")
plt.ylabel("Numero di Guariti")
plt.title("Confronto tra guariti post e senza somministrazione")

plt.show()
```
